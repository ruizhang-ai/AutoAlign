```python
from rdflib.term import Literal, URIRef
```

```python
from rdflib import Graph

# 读取两个文件并加载到Graph中
g1 = Graph().parse("DY-NB/yago.ttl", format="ttl")
g2 = Graph().parse("DY-NB-current/yago.ttl", format="ttl")
```

```python
# 找出两个Graph中相同和不同的triple
same_triples = set(g1) & set(g2)
different_triples = set(g1) ^ set(g2)

# 输出结果
print(f"{len(same_triples)} triples are the same.")
print(f"{len(different_triples)} triples are different.")
```

```python
# 从different_triples集合中提取所有Literal类型的object
literal_objects = sorted(
    {triple[2] for triple in different_triples if isinstance(triple[2], Literal)}
)

# 输出结果
print(f"{len(literal_objects)} literal objects found:")
for obj in literal_objects:
    print(obj)
```

```python
uri_objects = sorted(
    {triple[0] for triple in different_triples if isinstance(triple[0], URIRef)}
)

# 输出结果
print(f"{len(uri_objects)} URIRef objects found:")
for obj in uri_objects:
    print(obj)
```

```python
from collections import Counter
from rdflib.term import URIRef

# 从different_triples集合中提取所有URIRef类型的object
uri_objects = [
    str(triple[0]).split("/")[-1] for triple in different_triples if isinstance(triple[0], URIRef)
] + [
    str(triple[2]).split("/")[-1] for triple in different_triples if isinstance(triple[2], URIRef)
]

# 计算每个项出现的次数
counter = Counter(uri_objects)

# 找出所有重复的项
duplicate_items = set([x for x, count in counter.items() if count > 1])

# 输出结果
if len(duplicate_items) > 0:
    print(f"{len(duplicate_items)} duplicate items found:")
    for item in duplicate_items:
        print(item)
else:
    print("No duplicate items found.")
```

```python
from rdflib.term import URIRef

# 从different_triples集合中提取所有URIRef类型的object
uri_objects = [
    str(triple[0]) for triple in different_triples if isinstance(triple[0], URIRef)
] + [
    str(triple[2]) for triple in different_triples if isinstance(triple[2], URIRef)
]

# 找到前缀不同但是最后一部分相同的URIRef对象
prefixes = set([uri.rsplit("/", 1)[0] for uri in uri_objects])
```

```python
prefixes
```

```python
for suffix in set([uri.rsplit("/", 1)[-1] for uri in uri_objects]):
    unique_prefixes = set()
    for uri in uri_objects:
        if uri.rsplit("/", 1)[-1] == suffix:
            prefix = uri.rsplit("/", 1)[0]
            #print(uri, prefix)
            if prefix not in unique_prefixes:
                unique_prefixes.add(prefix)
    if len(unique_prefixes) > 1:
        print(f"URIRef objects with different prefixes but same suffix '{suffix}':")
        for prefix in unique_prefixes:
            print(f"  {prefix}{suffix}")
```
